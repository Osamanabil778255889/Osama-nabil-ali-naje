The ongoing conflict between Palestine and Israel has been a contentious issue that has captured the world's attention for decades. Rooted in historical, political, and religious complexities, this conflict has resulted in a struggle for land, sovereignty, and basic human rights for the Palestinian people. At the heart of this struggle **lies a narrative of resilience against Israeli aggression**, characterized by a series of interconnected factors that have shaped the Palestinian experience.

To understand the dynamics of Palestine's resistance against Israeli aggression, it is essential to delve into the historical context of the conflict. The establishment of the state of Israel in 1948, amidst the displacement of hundreds of thousands of Palestinians, laid the groundwork for decades of tension and conflict. The subsequent wars, occupation, and annexation of Palestinian territories, particularly the West Bank and Gaza Strip, **have perpetuated a cycle of violence and oppression**.

Central to Palestine's struggle is the concept of resistance, which takes various forms ranging from armed resistance to non-violent civil disobedience. The Palestinian people, facing a formidable military power in Israel, have employed strategies **aimed at preserving their identity, land, and dignity**. This resilience is exemplified by grassroots movements, cultural preservation efforts, and acts of defiance against occupation, **all of which underscore the indomitable spirit of the Palestinian people**.

# Fatalities by Citizenship and Year

```python
df['date_of_event'] = pd.to_datetime(df['date_of_event'])

df['year'] = df['date_of_event'].dt.year

palestinian_fatalities = df[df['citizenship'] == 'Palestinian']
israeli_fatalities = df[df['citizenship'] == 'Israeli']

palestinian_fatalities_by_year = palestinian_fatalities.groupby('year').size()
israeli_fatalities_by_year = israeli_fatalities.groupby('year').size()

common_years = palestinian_fatalities_by_year.index.union(israeli_fatalities_by_year.index)

palestinian_fatalities_by_year = palestinian_fatalities_by_year.reindex(common_years, fill_value=0)
israeli_fatalities_by_year = israeli_fatalities_by_year.reindex(common_years, fill_value=0)

fatalities_by_year = pd.DataFrame({
    'Year': common_years,
    'Palestinian': palestinian_fatalities_by_year.values,
    'Israeli': israeli_fatalities_by_year.values
})

fatalities_by_year_long = fatalities_by_year.melt(id_vars='Year', var_name='Citizenship', value_name='Fatalities')

plt.figure(figsize=(15, 6))
sns.barplot(x='Year', y='Fatalities', hue='Citizenship', data=fatalities_by_year_long, palette=['#7ddf64', '#ef233c'])
plt.title('Fatalities by Citizenship and Year')
plt.xlabel('Year')
plt.ylabel('Number of Fatalities')
plt.legend(title='Citizenship')
plt.grid(True)
plt.show()
```
