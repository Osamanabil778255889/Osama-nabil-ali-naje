name = 'group_a', 'group_b', 'group_c' value=1, 10, 100

```python
groups = ['group_a', 'group_b', 'group_c']
values = [1, 10, 100]
fig, axs = plt.subplots(1, 3, figsize=(10, 4))
axs[0].bar(groups, values)
axs[0].set_title('Bar Plot')
axs[1].scatter(groups, values)
axs[1].set_title('Scatter Plot')
axs[2].plot(groups, values)
axs[2].set_title('Line Plot')
fig.suptitle('Categorical Plotting')
plt.show()
```

```python
year = [1972, 1982, 1992, 2002, 2012]
e_india = [100.6, 158.61, 305.54, 394.96, 724.79]
e_bangladesh = [10.5, 25.21, 58.65, 119.27, 274.87]
plt.figure(figsize=(8, 5))
plt.plot(year, e_india, label='India', color='orange')
plt.plot(year, e_bangladesh, label='Bangladesh', color='green')
plt.xlabel('Years')
plt.ylabel('Power consumption in kWh')
plt.title('Electricity consumption per capita of India and Bangladesh')
plt.legend()
plt.grid(True)
plt.show()
```
